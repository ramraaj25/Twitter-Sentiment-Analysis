```python
import pandas as pd
import nltk
import pickle
import string
```

```python

# open cleaned dataset
pickle_in = open('cleaned_dataset.pickle', 'rb')
data = pickle.load(pickle_in)
pickle_in.close()
data.sample(frac=1).reset_index(drop=True)
```

```python
data = data[:15000]
print(data.head())
```

```python
# splitting up happy and sad tweets
happy_tweets = []
sad_tweets = []
for i in range(len(data['tweet'])):
    if data['mood'][i] == 0:    # 0 represents sad mood
        sad_tweets.append(data['tweet'][i])
    elif data['mood'][i] == 4:  # 4 represents happy mood
        happy_tweets.append(data['tweet'][i])

```

```python
# creating a set of all the words present in the dataset
words = set()

# 'tweet' is a list of words(strings)
for tweet in happy_tweets:
    words.update(tweet)
for tweet in sad_tweets:
    words.update(tweet)
```

```python

# function to generate features in the correct format to feed the data into the model
def generate_features(tweets, words, label):
    features = []
    i = 0
    for tweet in tweets:
        i += 1
        features.append(({word: (word in tweet) for word in words}, label))
        print(f'generating_features...{i}')

    return features
```

```python
# generate features for happy tweets
feature_set = []
feature_set.extend(generate_features(happy_tweets, words, "Happy"))

# splitting up train and test sets
train_set = feature_set[:len(feature_set) - 2500]
test_set = feature_set[len(feature_set) - 2500:]
j = len(feature_set)

# generate features for sad tweets
feature_set.extend(generate_features(sad_tweets, words, "Sad"))

# splitting test and train sets
train_set.extend(feature_set[j:len(feature_set) - 2500])
test_set.extend(feature_set[len(feature_set) - 2500:])


```

```python
# training the model on the train set
classifier = nltk.NaiveBayesClassifier.train(train_set)

#   calculating the accuracy percent of the model on the test set
acc = nltk.classify.accuracy(classifier, test_set) * 100
print(f"Classifier accuracy percent: {acc}")
```

```python
# saving the list of words to generate features for new tweets
print('saving words....')
pickle_words = open('words.pickle', 'wb')
pickle.dump(words, pickle_words)
pickle_words.close()
```

```python
# saving our model
print('saving model....')
pickle_classifier = open('NaiveBayesClassifier.pickle', 'wb')
pickle.dump(classifier, pickle_classifier)
pickle_classifier.close()
print('done')
```
